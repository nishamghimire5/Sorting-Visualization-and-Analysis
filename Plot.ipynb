```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('data.csv')
```

__All algorithms__

```python
plt.figure(figsize=(10, 7))
plt.plot(df["Input Size"], df["Merge Sort"], label = "Merge Sort", color = "pink")
plt.plot(df["Input Size"], df["Insertion Sort"], label = "Insertion Sort", color = "red")
plt.plot(df["Input Size"], df["Quick Sort"], label = "Quick Sort", color = "blue")
plt.plot(df["Input Size"], df["Bubble Sort"], label = "Bubble Sort", color = "purple")
plt.plot(df["Input Size"], df["Heap Sort"], label = "Heap Sort", color = "green")
plt.plot(df["Input Size"], df["Selection Sort"], label = "Selection Sort", color = "black")
plt.legend()
plt.title("Sorting Algorithms Comparision")
plt.xlabel("Input Size (n)")
plt.ylabel("Time (s)")
plt.show()
```
